```python
# Problem 1: Social Media – Post Engagement Analyzer
# Problem Statement
# Create a function to analyze engagement on a social media post.
# Rules
# Calculate total likes
# If Total Likes ≥ 1000 → Viral Post
# Otherwise → Normal Engagement
```

```python
def check_engagement(likes_list):
    # Calculate total likes using loop
    total_likes = 0
    for like in likes_list:
        total_likes = total_likes + like

    # Check if viral or normal
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)

# --- Calling the function ---
likes = [200, 350, 150, 200, 150]
check_engagement(likes)
```

```python
# Problem 3: Agriculture – Rainfall Adequacy Checker
# Problem Statement
# Create a function that analyzes seasonal rainfall data.
# Rules
# Calculate average rainfall
# Average ≥ required level → Adequate Rainfall
# Otherwise → Inadequate Rainfall
```

```python
def check_rainfall(rainfall_list, required_level):
    # Calculate total using loop
    total = 0
    for rain in rainfall_list:
        total = total + rain

    # Calculate average
    average = total / len(rainfall_list)

    # Check adequacy
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", average)
    print("Rainfall Status:", status)

# --- Calling the function ---
monthly_rain = [60, 80, 75, 70, 85, 65]
check_rainfall(monthly_rain, 70)
```

```python
# Problem 4: Social Media – Duplicate Account Detection
# Problem Statement
# Create a function to detect duplicate usernames in a system.
# Rules
# If duplicates exist → Flag accounts
# Else → Accounts are unique
```

```python
def detect_duplicates(usernames):
    # Convert list to set (set removes duplicates)
    unique_names = set(usernames)

    # If lengths differ → duplicates exist
    if len(unique_names) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

# --- Calling the function ---
users = ["alice", "bob", "alice", "charlie", "bob"]
detect_duplicates(users)
```

```python
# Problem 6: Agriculture – Premium Crop Price Filter
# Problem Statement
# Create a function that filters crop prices above a premium threshold.
# Rules
# Price > ₹2000 → Premium Crop
# Else → Regular Crop
```

```python
def filter_premium_crops(prices):
    # Filter prices above 2000
    premium = []
    for price in prices:
        if price > 2000:
            premium.append(price)

    print("Premium Crops:", premium)

# --- Calling the function ---
crop_prices = [1500, 2500, 1800, 3200, 900]
filter_premium_crops(crop_prices)
```

```python
# Problem 7: System Monitoring – Application Health Checker
# Problem Statement
# Create a function to evaluate application health based on error count.
# Rules
# Errors = 0 → Healthy
# Errors ≤ 5 → Minor Issues
# Errors > 5 → Critical Issues
```

```python
def check_health(errors):
    print("Error Count:", errors)

    # Check error levels
    if errors == 0:
        print("System Status: Healthy")
    elif errors <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")

# --- Calling the function ---
check_health(7)
check_health(0)
check_health(3)
```

```python
# Problem 8: Banking – Daily Transaction Limit Checker
# Problem Statement
# Create a function to validate daily transaction amount.
# Rules
# Daily limit = ₹50,000
# Amount within limit → Approved
# Amount exceeds limit → Rejected
```

```python
def check_transaction(amount):
    daily_limit = 50000    # daily limit is 50000

    print("Transaction Amount:", amount)

    # Check against limit
    if amount <= daily_limit:
        print("Transaction Status: Approved")
    else:
        print("Transaction Status: Rejected")

# --- Calling the function ---
check_transaction(60000)
check_transaction(30000)
```

```python
# Problem 9: E-Learning – Student Attendance Eligibility System
# Problem Statement
# Create a function to calculate attendance percentage and check exam eligibility.
# Rules
# Attendance ≥ 75% → Eligible
# Attendance < 75% → Not Eligible
```

```python
def check_attendance(attended, total_classes):
    # Calculate percentage
    percentage = (attended / total_classes) * 100

    print("Attendance Percentage:", percentage)

    # Check eligibility
    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")

# --- Calling the function ---
check_attendance(40, 50)
check_attendance(30, 50)
```

```python
# Problem 10: Smart Electricity Bill Analyzer
# Problem Statement
# You are building a smart electricity billing system for residential users.
# Create a function that calculates the electricity bill and classifies usage level based on units consumed.
# Billing Rules
# First 100 units → ₹3 per unit
# Next 100 units (101–200) → ₹5 per unit
# Above 200 units → ₹7 per unit
# Usage Classification
# Total bill < ₹500 → Low Usage
# ₹500 – ₹1500 → Moderate Usage
# ₹1500 → High Usage
```

```python
def calculate_bill(units):
    bill = 0

    # Slab 1: First 100 units → Rs.3 per unit
    if units <= 100:
        bill = units * 3

    # Slab 2: 101 to 200 units → Rs.5 per unit
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)

    # Slab 3: Above 200 units → Rs.7 per unit
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Classify usage
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    # Display results
    print("Units Consumed:", units)
    print("Total Bill: ₹" + str(bill))
    print("Usage Level:", status)

# --- Calling the function ---
calculate_bill(90)
print("---")
calculate_bill(180)
print("---")
calculate_bill(350)
```
